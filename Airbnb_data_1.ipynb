```python

from sklearn.model_selection import KFold
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
```

```python
df1 = df[[ "beds", "review_scores_value", "price"]]
```

```python
df1.head(3)
```

```python
def convert_price_to_num(p):
    p_new =p.replace("$","").replace(",","")
    p_new = p_new[:-3] # remove .00 at the end
    return (int(p_new))
```

```python
df1["price"].map(convert_price_to_num)
```

```python
df1["price"]=df1["price"].map(convert_price_to_num)
```

```python
df1[pd.isnull(df1["beds"])]
```

```python
df1.fillna(0, inplace=True)
# alternatively
# df1 = df1.fillna(0)

```

```python
# just for checking whether fillna has worked
df1[pd.isnull(df1["beds"])]

```

```python
X = df1.drop("price", axis=1)
y = df1["price"]
```

```python
k = 5
kf = KFold(n_splits=k, random_state=None)
model = LogisticRegression(solver= 'liblinear')
acc_score = []
for train_index , test_index in kf.split(X):
    X_train , X_test = X.iloc[train_index,:],X.iloc[test_index,:]
    y_train , y_test = y[train_index] , y[test_index]

    
```

```python
model.fit(X_train,y_train)
pred_values = model.predict(X_test)
```

```python
acc = accuracy_score(pred_values , y_test)
acc_score.append(acc)
```

```python
acc_score
```

```python
avg_acc_score = sum(acc_score)/k
```

```python
print('accuracy of each fold - {}'.format(acc_score))
print('Avg accuracy : {}'.format(avg_acc_score))
```

```python


result = cross_val_score(model , X, y, cv = kf)
 
print("Avg accuracy: {}".format(result.mean()))
```
